Import the necessary packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import and check out the data 

```python
oil_dataset = pd.read_csv('cleaned_datasets/date_oil_fuel_tax.csv')
```

```python
print(oil_dataset.head())
```

```python
print(oil_dataset.info())
```

First we need to convert the Date column to date format

```python
oil_dataset['Date'] = pd.to_datetime(oil_dataset['Date'], dayfirst=False)
```

Now we'll need to put the dataset in ascending order

```python
oil_dataset.sort_values(by='Date', ascending=True, inplace=True)
oil_dataset
```

Lets also add a column to the dataset showing the total tax paid per litre

```python
oil_dataset['Total tax %'] = (oil_dataset['Dutyrate'] + oil_dataset['vat'])/100
tax_on_petrol = oil_dataset['Petrol(pence/litre)'] * oil_dataset['Total tax %']
tax_on_diesel = oil_dataset['Diesel(pence/litre)'] * oil_dataset['Total tax %']
```

Now let's draw a linegraph to show how the price changes

```python
plt.figure(figsize=(16,9))
plt.plot(oil_dataset['Date'], oil_dataset['Barrel_oil_£'], label='Barrel of Oil in £s', color='blue')
plt.plot(oil_dataset['Date'], oil_dataset['Petrol(pence/litre)'], label='Petrol, Pence/litre', color='green')
plt.plot(oil_dataset['Date'], oil_dataset['Diesel(pence/litre)'], label='Diesel Pence/litre', color='black')
plt.legend()
plt.grid()
plt.show()
```

Graph inaccurate, need to convert cost of oil into pence/litre,
42 gallons in a barrel, 4.54609 litres in a gallon, 

```python
def barrel_to_pence_litre(df):
    new_price = (df * 42 * 4.54609)/ 100
    return new_price
oil_dataset['oil pence/litre'] = barrel_to_pence_litre(oil_dataset['Barrel_oil_£'])
```

```python
oil_dataset
```

Now lets look at the prices where they are all measured in pence per litre

```python

plt.figure(figsize=(16,9))
plt.plot(oil_dataset['Date'], oil_dataset['oil pence/litre'], label='oil pence/litre', color='blue')
plt.plot(oil_dataset['Date'], oil_dataset['Petrol(pence/litre)'], label='Petrol, Pence/litre', color='green')
plt.plot(oil_dataset['Date'], oil_dataset['Diesel(pence/litre)'], label='Diesel pence/litre', color='black')
plt.plot(oil_dataset['Date'], tax_on_petrol, color='red', label='Petrol Tax', alpha=0.5)
plt.plot(oil_dataset['Date'], tax_on_diesel, color='pink', label='Diesel Tax', alpha=0.5)
plt.legend()
plt.xlabel('Date')
plt.ylabel('Pence per Litre')
plt.grid()
plt.show()
```

This gives us a clearer picture as all units are now the same, going from the graph you would have to say that the price of oil does directly affect the price of petrol and diesel.  This graph also shows us the amount paid in tax for each litre of fuel, which is staggeringly high

# Changes in Price

```python
oil = oil_dataset[['Date', 'oil pence/litre']].copy()
petrol = oil_dataset[['Date', 'Petrol(pence/litre)']].copy()
diesel = oil_dataset[['Date', 'Diesel(pence/litre)']].copy()
petrol
```

Now we've seperated the datasets, lets figure out the percentage change from the row before

```python
oil['Oil Percent Change'] = oil['oil pence/litre'].pct_change() * 100
petrol['Petrol Percent Change'] = petrol['Petrol(pence/litre)'].pct_change() * 100
diesel['Diesel Percent Change'] = diesel['Diesel(pence/litre)'].pct_change() * 100
```

Lets plot these changes on another line graph

```python
plt.figure(figsize=(16,9))
plt.plot(oil['Date'], oil['Oil Percent Change'],color='blue', alpha=0.2, label='Oil')
plt.plot(petrol['Date'], petrol['Petrol Percent Change'],color='green', label='Petrol')
plt.plot(diesel['Date'], diesel['Diesel Percent Change'], color='black', alpha=0.4, label='Diesel')
plt.xlabel('Year')
plt.ylabel('% change in price')
plt.legend()
plt.grid()
plt.show()
```

As this graph looks quite messy, we'll summarise it better by grouping the dates by year and summing the changes in price

```python
yearly_oil = oil.groupby(oil['Date'].dt.year)['Oil Percent Change'].agg(['sum', 'mean', 'max'])
yearly_petrol = petrol.groupby(petrol['Date'].dt.year)['Petrol Percent Change'].agg(['sum', 'mean', 'max'])
yearly_diesel = diesel.groupby(diesel['Date'].dt.year)['Diesel Percent Change'].agg(['sum', 'mean', 'max'])

```

```python
plt.figure(figsize=(16, 9))
plt.plot(yearly_oil['sum'], color='blue', label = 'Oil')
plt.plot(yearly_petrol['sum'], color='green', label='Petrol')
plt.plot(yearly_diesel['sum'], color='black', label='Diesel')
plt.xticks([x for x in range(2002, 2023)])
plt.legend()
plt.grid()
plt.show()
```
